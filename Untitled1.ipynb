```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,np.pi*2,20).round(2).tolist()
y = np.sin(x).round(2).tolist()
print(list(zip(x,y)))

fig, ax = plt.subplots()
ax.plot(x, y, marker='o', linestyle='-')
# ax.scatter(x, y, color="red", s=15)

```
